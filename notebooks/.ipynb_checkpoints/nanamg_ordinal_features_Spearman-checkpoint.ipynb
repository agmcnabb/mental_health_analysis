# Spearman´s Rank Correlation applied on Ordinal features

```python
import pandas as pd
import os
import pathlib
import matplotlib.pyplot as plt
import json
import numpy as np

import re
from scipy.stats import spearmanr

import seaborn as sns
import matplotlib.pyplot as plt
```

## Extracting the names of the features asigned to Ordinal

```python
notebook_dir = pathlib.Path().resolve()
# Path to the JSON file
filepath_json = os.path.join(notebook_dir,'../raw_data/feature_lists_dict.json')

with open(filepath_json, 'r') as file:
    data_list = json.load(file)
```

```python
# Extracting JSON into our features list
DS_numerical = []
DS_cat_nominal = []
DS_cat_ordinal_lowest_num_is_highest_value = []
DS_cat_ordinal_highest_num_is_highest_value = []
DS_cat_nominal_binary_Y1_N2 = []
DS_cat_nominal_binary_Y1_N0 = []

for key, value in data_list.items():
    if key == 'DS_numerical':
        DS_numerical.extend(value)
    elif key == 'DS_cat_nominal':
        DS_cat_nominal.extend(value)
    elif key == 'DS_cat_ordinal_lowest_num_is_highest_value':
        DS_cat_ordinal_lowest_num_is_highest_value.extend(value)
    elif key == 'DS_cat_ordinal_highest_num_is_highest_value':
        DS_cat_ordinal_highest_num_is_highest_value.extend(value)
    elif key == 'DS_cat_nominal_binary_Y1_N2':
        DS_cat_nominal_binary_Y1_N2.extend(value)
    elif key == 'DS_cat_nominal_binary_Y1_N0':
        DS_cat_nominal_binary_Y1_N0.extend(value)
```

#### Replacing -1, -8 and -9  for most frequent value

## Extracting y

```python
smfq_variables = [
    'FCMDSA00',
    'FCMDSB00',
    'FCMDSC00',
    'FCMDSD00',
    'FCMDSE00',
    'FCMDSF00',
    'FCMDSG00',
    'FCMDSH00',
    'FCMDSI00',
    'FCMDSJ00',
    'FCMDSK00',
    'FCMDSL00',
    'FCMDSM00'
]
```

```python
# relative path to data_raw
raw_features_path_4 = os.path.join(notebook_dir,'../raw_data/accepted_features.csv')
y = pd.read_csv(raw_features_path_4, index_col=None)
```

```python
y = y[(y['Variable name'].isin(smfq_variables))]
```

```python
y_features = y['Variable label'].tolist()
```

```python
y_features = list(map(str.strip, y_features))
```

```python
y_features
```

##### Removing y_features from the dataframe

```python
all_present = all(col in data.columns for col in y_features)

if all_present:
  print("All columns from 'y_features' are present in the DataFrame.")
else:
  # Optionally, find missing columns
  missing_cols = [col for col in y_features if col not in data.columns]
  print(f"Following columns from 'y_features' are missing: {', '.join(missing_cols)}")
```

```python
data = data.drop(y_features, axis=1)
```

## Spearmans´s Rank Corr Coeficient

```python
# calculate the spearman's correlation between two variables. For p-value > 0.05 we fail to reject H0, samples are uncorrelated.
# if p-value < 0,05, reject H0, samples are correlated
```

```python
spearman_corr = data.corr(method='spearman')
```

```python
# Correlation values 
spearman_corr
```

```python
# p-values
p_values = pd.DataFrame(index=data.columns, columns=data.columns)
for col1 in data.columns:
    for col2 in data.columns:
        rho, p_value = spearmanr(data[col1], data[col2])
        p_values.loc[col1, col2] = round(p_value, 3)
```

```python
p_values
```

```python
p_values['ActivGrid: Go to the cinema?'].unique()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(p_values, cmap='coolwarm', annot=True)
plt.title("Spearman Correlation P-Values (Threshold: 0.05)")
plt.xlabel("Columns")
plt.ylabel("Columns")
plt.show()
```
